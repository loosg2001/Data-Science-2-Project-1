```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ssl
import urllib.request
import certifi
from ucimlrepo import fetch_ucirepo
```

```python
_real_urlopen = urllib.request.urlopen

def _patched_urlopen(url, *args, **kwargs):
    if 'context' not in kwargs:
        kwargs['context'] = ssl.create_default_context(cafile=certifi.where())
    return _real_urlopen(url, *args, **kwargs)

urllib.request.urlopen = _patched_urlopen

# fetch dataset 
auto_mpg = fetch_ucirepo(id=9)
```

```python
#data (as pandas dataframes)
x = auto_mpg.data.features
y = auto_mpg.data.targets

df = pd.concat([x, y], axis=1)

print(f'df.head() = \n{df.head()}')
print(f'\ndf.shape = \n{df.shape}')
print(f'\ndf.columns = \n{df.columns.tolist()}')
```

```python
print(df.isnull().sum())
df = df.dropna()
print(df.isnull().sum())
```

```python
df.insert(loc=0, column='intercept', value=1)

print(f'df.head() = \n{df.head()}')
print()
print(df.isnull().sum())



df.to_csv("cleaned_auto_mpg_with_intercept.csv", index=False)
print("Successfully generated cleaned_auto_mpg_with_intercept.csv")
```

```python
df.drop('intercept', axis=1).hist(figsize=(12,8), bins=20)
plt.tight_layout()
plt.savefig('hisograms.png',dpi = 300,bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(16,10))
for i, col in enumerate(['displacement','horsepower','weight','acceleration']):
    plt.subplot(2,2,i+1)
    sns.boxplot(y=df[col])
    plt.title(col, fontsize=14)
    plt.ylabel(col, fontsize=12)

plt.tight_layout(pad = 3.0)
plt.savefig('boxplots.png',dpi=400,bbox_inches='tight')
plt.show()
```

```python
features = ['displacement','cylinders','horsepower','weight','acceleration','model_year','origin']

plt.figure(figsize=(16,12))

for i, col in enumerate(features):
    plt.subplot(4,2,i+1)
    sns.scatterplot(x=df[col], y=df['mpg'])
    plt.title(f"{col} vs mpg")

plt.tight_layout(pad=3.0)
plt.savefig('scatterplots.png',dpi=400,bbox_inches='tight')
plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(corr , annot = True , cmap = 'coolwarm')

plt.tight_layout()
plt.savefig('correlation_heatmap.png',dpi = 300,bbox_inches='tight')
plt.show()
```
